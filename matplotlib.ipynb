```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python


x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]


# plt.plot(x, y)
plt.plot(x,y,color='black',marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = ["python","c","c++","java"]
y = [90,80,75,85]
plt.bar(x,y)
plt.show()
```

```python
x = ["python","c","c++","java"]
y = [90,80,75,85]
colors = ['r','b','m','black']
plt.barh(x,y,color=colors)
plt.show()
```

```python
x = ["python","c","c++","java"]
y = [90,80,75,85]
plt.bar(x,y,width=0.2,color='black')
plt.xlabel("language",fontsize=6)
plt.ylabel("No",fontsize=6)
plt.title("Plot")
plt.show()
```

```python
x = ['Category A','Category B','Category C']
sizes = [30,45,25] 

plt.pie(sizes,labels=x)
plt.title('Pie chart')
plt.show()
```

```python
data = np.random.randn(1000)

plt.hist(data , bins=30 ,edgecolor='black')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
x = ["python","c","c++","java"]
y=[90,80,75,85]
color=["r","g","y","orange"]
size=[100,200,300,400]
# plt.scatter(x,y)
plt.scatter(x,y,c=color,s=size,alpha=0.99,marker="*",edgecolor="black")
# t=plt.colorbar()
# plt.title("Color Bar")
plt.title("Scatter plot",fontsize=18)
plt.xlabel("Programming language",fontsize=15)
plt.ylabel("Popularity ",fontsize=15)
plt.show() 
```

```python
x = [1,2,3,4,5]
y1 = [10,15,25,30,35]
y2 = [5,10,20,25,30]

plt.fill_between(x,y1,y2,color = 'blue')
plt.title ("Area Chart")
plt.xlabel('X-axis')
plt.ylabel('Y-axis') 
plt.show() 
```

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.title("STEP")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid()
plt.step(x,y,marker="*")
plt.show() 
```
